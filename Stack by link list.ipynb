```python
# 6. Size

class Node:
    def __init__(self,item=None, next=None):
        self.item=item
        self.next=next
        
class Stack:
    def __init__(self):
        self.start=None
        self.item_count=0
        
    def is_empty(self):
        return self.start==None
    
    def push(self,data):
        n=Node(data,self.start) # node mein data aur start ki value( whatever null or next node address)
        self.start=n
        self.item_count+=1   
        
    def pop(self):
        if not self.is_empty():
            data=self.start.item
            self.start=self.start.next
            self.item_count-=1
            return data
        else: 
            raise IndexError("Stack is empty")
        
    def peek(self):
        if not self.is_empty():
            return self.start.item
        else: 
            raise IndexError("Stack is empty")
        
    # solution here
    def size(self):
        return self.item_count
    
s1=Stack()
s1.push(10)
print("Peek element is ",s1.peek())
s1.push(20)
print("Peek element is ",s1.peek())
s1.push(30)
print("Peek element is ",s1.peek())
s1.pop()
print("Peek element is ",s1.peek())
print("Size of stack ",s1.size())
s1.push(40)
print("Peek element is ",s1.peek())
print("Size of stack ",s1.size())

        
```
