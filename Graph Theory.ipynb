```python

# A graph consists of two things:
# 1. A set "V" of elements called Node(Vertices) # A tree is a special case of Graph 
# 2. A set "E" of edges such that each edge e in "E" is identified as
# unique (unordered) pair [u,v] of nodes in v , denoted by 
# e = [u,v] 
# We indicate the parts of the graph by wirting G=(v,e)

# Adjacent nodes:- 
# If e=[u,v] , then u and v are called adjacent or neighbors

# Degree of Node 
# The degree of node u, written deg(u), is the number of edges 
# containing u
#** If deg(u)=0 , then u is called isolated node

# Path
# A path of length n from a node "U" to a node "V"
# is defined as a sequence of n+1 nodes

# P=(v0,v1,v2,..v3)
# The path is said to be closed if v0 = vn

# Simple and complex path
# The path is said to be simple if all the nodes are distinct
# with the exception that v0 may equal to vn 
# otherwise it is complex path

# Connected Graph:- A graph is said to be connected if there
# is a path between any two of its nodes

# Tree Graph:- A connected graph "T" without any cycles is tree graph
# or free tree or simply a tree

# ** This means is particular, that is a unique simple path
# "P"  between any two nodes u and v 

# Labelled Graph :- A graph is to be labelled if its edges
# are assigned data

# Weighted Graph :- 
# A graph "G" is said to be weighted if each edge "e" in "G"
# is assigned a non-negative numerical value W(e)
# called the weight or length of e

# Loop :- An edge is called loop if it has identical end point
# e=[u,v]

# Directed Graph:- A directed graph "G" also called digraph
# is same as multigraph except that each edge e is assigned a 
# direction e=(u,v)

# Representation of a Graph
# 1.Adjacency Matrix Representation
# 2.List repsentation

# Adjacency matrix representation:- 
# Suppose G is a simple graph with m nodes
# and suppose the nodes of G have been ordered and 
# are called v1,v2,v3,..vn. 
# Then the adjacency matrix A = (aij) of the graph G
# is the  main matrix defined as follows : -
# aij=1 if vi is adjacent of vj
  #   0 otherwise

# Adjacency list comprehension
# The adjacency list stores information about only those edges that exists
# v0,v1,v2,..vn
# Teh adjacency list contains a directory(dict) containing
# adjacency list for each vertex

```
