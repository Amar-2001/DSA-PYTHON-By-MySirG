```python
# Node Class, BST Class, insert()+rinsert(), search()+rserach(), Inorder(), Preorder(),Postorder(), Min_value(), Max_Value(),
#  Delete()+rdelete(), Size()

class Node:
    def __init__(self,item=None,left=None,right=None):
        self.item=item
        self.left=left
        self.right=right

class BST:
    def __init__(self):
        self.root=None

    def insert(self,data):
        self.root=self.rinsert(self.root,data) # calling self.rinsert and storing in self.root

    #  rinsert function bana rahe
    def rinsert(self,root,data):
        if root is None:
            return Node(data) # return kar do Node jisme argument data ho baki left aur right None hi ho
                                # Node ke argument mein sirf data pass kiye left aur right nahi toh 
                                # by default wo None hi hoga

        if data<root.item: # agar data less than hai root se toh left subtree mein add ho data item
            root.left= self.rinsert(root.left,data)

        elif data>root.item:# agar data more than hai root se toh right subtree mein add ho data item
            root.right= self.rinsert(root.right,data)
        else:
            return root
            
    # To search
    def search(self,data): 
        return self.rsearch(self.root,data) # recursivley search data 
    
    def rsearch(self,root,data): # recursion
        if root is None or root.item==data:  # agar root none hai ya root item data se match karta toh return root ok
            return root
        if data<root.item: # data < root toh left mein seach recursively
            return self.rsearch(root.left,data)
        else: # data > root toh right mein seach recursively
            return self.rsearch(root.right,data)
        

    #Inorder (left,root,right)
    def inorder(self): 
        result=[] # result list mein data append(rakhna) karna hai 
        self.rinorder(self.root,result) # calling rinorder and passing (root and result list )
        return result # last mein result return kar do
    def rinorder(self,root,result): # recursive function of inorder
        if root: # agar root empty nahi hua toh (shortcut trick of function )
            self.rinorder(root.left,result)  # left mein recursivley call
            result.append(root.item) # result mein root ka item append (rakho)
            self.rinorder(root.right,result) # right mein recursivley call
        #** return karane ki koi jarurat nahi kyu ki result toh same list ko refer kar raha******************

    # Preorder(root,left,right)
    def preorder(self):
        result=[]
        self.rpreorder(self.root,result)
        return result
    def rpreorder(self,root,result):
        if root:
            result.append(root.item)
            self.rpreorder(root.left,result)
            self.rpreorder(root.right,result)
#** return karane ki koi jarurat nahi kyu ki result toh same list ko refer kar raha******************

    #Post order (left,right,root)
    def postorder(self):
        result=[]
        self.rpostorder(self.root,result)
        return result
    def rpostorder(self,root,result):
        self.rpostorder(root.left,result)
        self.rpostorder(root.right,result)
        result.append(root.item)
#** return karane ki koi jarurat nahi kyu ki result toh same list ko refer kar raha******************

    # To find min value ,we have to go left
    def min_value(self,temp):
        current=temp # current mein temp rakh do 
        while current.left is not None:
            current=current.left
        return current.item
    
    def max_value(self,temp):
        current=temp
        while current.right is not None:
            current=current.right
        return current.item
    
    # To delete

    def delete(self,data):  
        self.root=self.rdelete(self.root,data) # updated tree ke root node ka address isliye root(self.rot) mein store kar rahe

    def rdelete(self,root,data):
        if root is None:# CAse 1: when no child  (only next two line code is for this after that it is for sigle child)
            return root # agar root none hai toh return karo root(yaha tum none bhi return kar sakte ho root initially none hi hai)
# To find data whether it is in left or right         
        if data<root.item:
            root.left=self.rdelete(root.left,data)
        elif data>root.item:
            root.right=self.rdelete(root.right,data)

        # agar item mil gaya tab
        else:
    #  Case 2.to delete tree data having single child( whethere they are in left or right)

            # if the data we want to delete having children on the right side(means left side none) then return right side
            if root.left is None: # root ka left mein koi subtree nahi hai 
                return root.right # root ke right wale ko return kar do 
                                  # ye store ho jayega delete kiye huye item ke place par (due to recursion last position)
                                  # in copy example see line 98
            # if the data we want to delete having children on the left(means right side none) then return right side
            elif root.right is None:
                return root.left
            
    # Case 3. agar root ke two child ho then 
            root.item=self.min_value(root.right) # successor se replace (root ke right ke min value se replace)
            self.rdelete(root.right,root.item)
        return root # for case 3 return root 

        #     root.item=self.max_value(root.left) # predecessor se replace(root lke left max value se replace)
        #     self.rdelete(root.left,root.item)
        # return root
    
    # **************************************************
    # Now size
    def size(self):
        return len(self.inorder()) # inorder function ki lenght se size pata chal jayega 
                                # inorder function also give output in ascending order

        
bst=BST()
bst.insert(10)
bst.insert(20)
bst.inorder()


```
